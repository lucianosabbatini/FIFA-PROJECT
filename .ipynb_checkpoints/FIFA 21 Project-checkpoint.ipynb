#### Making columns names standard lower case

```python
#making all columns names lower case
cols = []
for column in data.columns:
    cols.append(column.lower())
cols
data.columns = cols
```

```python
data=data.drop(['id', 'age', 'weight', 'weight', 'growth', 'value (euros)', 'wage (euros)'], axis=1)
```

```python
data.describe(include=[np.object]).T
```

```python
data=data.drop(['name', 'nationality', 'club', 'bp', 'position', 'team & contract', 'height', 'weight', 'foot', 'joined', 'release clause', 'w/f', 'sm', 'a/w', 'd/w', 'ir', 'hits'], axis=1)
```

## Cleaning

```python
#Checking and dropping all for duplicates. There are none
data.duplicated()
data=data.drop_duplicates()
```

#### Dealing with null values

```python
#Checking the Nulls for numeric data
#data_na_n=data.select_dtypes(include=[np.number])
#data_na_n.isna().sum()
```

```python
#Some columns have the exact same amount of Nulls, maybe it´s easier to drop them as they represent 0.15% of the dataset
```

```python
#Checking the Nulls for categorical data
data_na_o=data.select_dtypes(include=[np.object])
data_na_o.isna().sum()
```

```python
#This column has 100% of null values, therefore, useless to the dataframe, hence we are dropping it.
data=data.drop(['loan date end'], axis=1)
```

```python
#Column
```

```python
nulls_df = pd.DataFrame(round(data.isna().sum()/len(data),4)*100)
```

```python
nulls_df
```

```python
#We noticed that several columns had 41 nulls. We checked if they refered for the same rows. 
data2 = pd.DataFrame(data,columns=['volleys','curve', 'balance', 'jumping', 'sliding tackle', 'vision'])

nan_values = data2[data2.isnull().any(axis=1)]

print (nan_values)
```

```python
#They represent 0.15% of the data so we are just dropping them
data=data[data['volleys'].isna()==False]
data=data[data['curve'].isna()==False]
data=data[data['balance'].isna()==False]
data=data[data['jumping'].isna()==False]
data=data[data['sliding tackle'].isna()==False]
data=data[data['vision'].isna()==False]
```

```python
data['composure']=data['composure'].fillna(data['composure'].median())
```

```python
data['club']=data['club'].fillna('Unknown')
data['position']=data['position'].fillna('Unknown')
data['a/w']=data['a/w'].fillna('Unknown')
data['d/w']=data['d/w'].fillna('Unknown')
```

## Cleaning Column by column

#### The FIFA 

```python
#data['position'].unique()
#this code returns a huge amount of unique values and there is a column that sums up those positions (column called bp). 
# So we are dropping the column position and naming the column bp as position.
data=data.drop(['position'], axis=1)
```

```python
#getting rid of the string object LBS to be able to change the column to numeric
data['weight'] = data['weight'].str.replace('lbs', '')
```

```python
#Now converting the column to numeric
data['weight'] = pd.to_numeric(data['weight'], errors='coerce')
```

```python
data['value'].unique()
```

```python
data['value'].value_counts()
```

```python
data['value'] = data['value'].str.replace('€', '')
```

```python
data.loc[data['value'].str.endswith('M') & ~data['value'].isnull(), 'value'] = data.loc[data['value'].str.endswith('M') & ~data['value'].isnull(), 'value'].str.replace('M', '').astype(float) * 1000000
```

```python
data.loc[data['value'].str.endswith('K') & ~data['value'].isnull(), 'value'] = data.loc[data['value'].str.endswith('K') & ~data['value'].isnull(), 'value'].str.replace('K', '').astype(float) * 1000
```

```python
data['value'] = data['value'].astype(int)
```

```python
data['value'].dtype
```

```python
#Doing the same to the column Wage
data['wage'].unique()
```

```python
data['wage'] = data['wage'].str.replace('€', '')
```

```python
data.loc[data['wage'].str.endswith('K') & ~data['wage'].isnull(), 'wage'] = data.loc[data['wage'].str.endswith('K') & ~data['wage'].isnull(), 'wage'].str.replace('K', '').astype(float) * 1000
```

```python
data['wage'] = data['wage'].astype(int)
```

```python
#Changing the name of these two columns to point out which currency and the ova to Overall Rating
data = data.rename(columns={'wage': 'wage (euros)'})
data = data.rename(columns={'value': 'value (euros)'})
data = data.rename(columns={'ova': 'overall rating'})
```
